# windows 단축키
- windows + e : 파일 탐색기
- windows + R : 실행창
- windows + R + cmd : 명령 프롬프트
- 프롬프트에는 기존에 있던 설정만 반영됨 , 설정을 변경했고 그를 확인하려면 기존의 cmd는 닫고 다시 열어야한다.
- python interpreter은 꺽새 3개 >>>다
- exit() 또는 cntrl z로 나올 수 있다.
- python -V : 현재 python의 버전을 보여줌 
- set python_home : 현재 python의 set과 경로를 보여줌
- 여러개의 python을 사용하기 위해서 환경변수를 설정한다.
- set path 하면 복잡한 경로가 나오는데, 이는 기존에 있던 경로로 없애면 안된다
- 설정한 path를 앞에 언급한 경로의 가장 앞에 위치시켜야한다. 그럼 그 버전이 먼저 우선 실행된다.
- anaconda의 python과 python은 별개로 생각한다. (아나콘다는 자체의 가상환경이 있다)


- 최신버전뿐만아니라 다른 버전을 설치하고 싶은경우 (다양한 버젼, embedded, installer, help file 등)

# 설치

- embedded : 작은소형전자기기 (라즈베리파이)를 위한 버젼, light version
- 라즈베리파이 안에 python2, python3 있음 따로 설치 필요 없음.
- help file : api 다큐멘트, 개발자될 사람이면 무조건 필독 
- installer : 보통설치파일(컴퓨터)
    - 메모리 4기가가 안되는 32bit설치 권장
    - 64bit 메모리 4기가 이상 인식 가능, 따라서 운영체제의 bit + 메모리 용량을 고려해서 설치를 한다.
    - 64bit는 32,64 모두 호환된다.
    - 32bit는 x86으로 적혀있다.
